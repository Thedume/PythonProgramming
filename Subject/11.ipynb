# 무 정보(radishsurvey.txt)데이터 투표자 분리

```python
counts = {}
voted = []
with open("../data/radishsurvey.txt") as f:
    for line in f:
        line = line.strip()

        name, vote = line.split(" - ")
        name = name.strip().capitalize().replace("  ", " ")
        if name in voted:
            # print("중복")
            continue
        voted.append(name)
        vote = vote.strip().capitalize().replace("  ", " ")
        if not vote in counts:
            counts[vote] = [name]
        else:
            counts[vote].append(name)

    for vote in counts:
        counts[vote].sort()

counts = dict(sorted(counts.items()))
# for name in counts:
#     print(f"{name} : {counts[name]}")

with open("../data/mySorting_radishsurvery.txt", 'a') as f:
    for vote in counts:
        f.write(f"###{vote} : ")
        for name in counts[vote]:
            if name == counts[vote][-1]:
                f.write(f"{name}.")
            else:
                f.write(f"{name}, ")

        f.write("\n\n")
```
