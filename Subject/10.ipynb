
출발 공항이나 도착 공항이 한국인 경우의 데이터 분석

```python
import csv
from geopy.distance import great_circle
import matplotlib.pyplot as plt

airport_KR = []
latitudes = {}
longitudes = {}
with open("C:\\Users\\JinYoung\\Desktop\\GIT\\PythonProgramming\\data\\airport.dat", encoding="UTF-8") as f:
    for row in csv.reader(f):
        if row[3] == "South Korea":
            airport_KR.append(row[0])

            latitudes[row[0]] = float(row[6])
            longitudes[row[0]] = float(row[7])

# print(airport_KR)
distance = []

with open("C:\\Users\\JinYoung\\Desktop\\GIT\\PythonProgramming\\data\\routes.dat", encoding="UTF-8") as f:
    for row in csv.reader(f):
        if row[3] in airport_KR or row[5] in airport_KR:
            # print(row[2], row[4])
            if row[3] in latitudes and row[5] in latitudes:
                s_lat = latitudes[row[3]]
                s_lon = longitudes[row[3]]
                d_lat = latitudes[row[5]]
                d_lon = longitudes[row[5]]
                distance.append(great_circle((s_lat,s_lon), (d_lat,d_lon)).km)


plt.hist(distance, 100)
plt.xlabel("Distance(km)")
plt.ylabel("Number of Flights")
```
